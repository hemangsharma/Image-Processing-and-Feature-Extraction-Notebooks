# Image Processing

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2  # For additional image processing

# Step 1: Load and Display the Image
# Load an image from file and display it
img_path = 'sample.png'  # Replace with your image path
original_img = Image.open(img_path)
plt.figure(figsize=(5, 5))
plt.title("Original Image")
plt.imshow(original_img)
plt.axis('off')
plt.show()

# Step 2: Convert to Grayscale

# Convert the image to grayscale for simplicity
gray_img = original_img.convert("L")
plt.figure(figsize=(5, 5))
plt.title("Grayscale Image")
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()

# Step 3: Resize the Image (Optional)
# Resize the image to a standard size (e.g., 28x28 pixels) to reduce data
resized_img = gray_img.resize((28, 28))
plt.figure(figsize=(5, 5))
plt.title("Resized Image (28x28)")
plt.imshow(resized_img, cmap='gray')
plt.axis('off')
plt.show()

# Step 4: Flatten the Image
# Convert the resized image to a NumPy array and flatten it
img_array = np.array(resized_img)
flattened_img_array = img_array.flatten()
print("Flattened Image Array:")
print(flattened_img_array)

# Step 5: Normalize Pixel Values
# Normalize pixel values to be between 0 and 1
normalized_img_array = flattened_img_array / 255.0
print("Normalized Flattened Image Array:")
print(normalized_img_array)

# Step 6: Visualize the Image Data as an Array
# Display the 28x28 pixel values for a grayscale image
plt.figure(figsize=(5, 5))
plt.title("Pixel Values (28x28)")
plt.imshow(img_array, cmap='gray')
plt.axis('off')
plt.colorbar()
plt.show()

# Step 7: Reshape and Visualize for Model Input
# Reshape the normalized data for machine learning model input (e.g., (1, 784) for 28x28)
model_input = normalized_img_array.reshape(1, -1)
print("Model Input Shape:", model_input.shape)
```
